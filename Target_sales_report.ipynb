```python
# Define the function to extract the variant
def get_variant(row):
    item_type_name = row['Item Type Name']
    item_sku_code = row['Item SKU Code']
    
    # Check the Item Type Name first
    if isinstance(item_type_name, str):  # Check if the value is a string
        if 'Ergo' in item_type_name:
            return 'Ergo'
        elif 'Ortho' in item_type_name:
            return 'Ortho'
        elif 'Antigravity' in item_type_name:
            return 'Antigravity'
        elif 'Wave' in item_type_name:
            return 'Wave'
    
    # If Item Type Name is not available or doesn't match, check the Item SKU Code
    if isinstance(item_sku_code, str):  # Check if the value is a string
        if item_sku_code.startswith('OB'):
            return 'Ortho'
    
    # Default to 'Unknown' if no conditions are met
    return 'Unknown'

# Apply the function to create the new column
data['Variant'] = data.apply(get_variant, axis=1)
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Border, Side, Alignment, Font

# Use the selected report date from Cell 1
report_date = pd.Timestamp(report_date)  # Ensure report_date is a Timestamp

# Get the first and last day of the current month
first_day = report_date.replace(day=1)
last_day = (first_day + pd.DateOffset(days=32)).replace(day=1) - pd.DateOffset(days=1)

# Generate a list of dates for the current month
dates = pd.date_range(start=first_day, end=last_day).to_list()

# Facility names and their corresponding labels
facility_names = ["FLO SAKINAKA", "FLO VASAI", "FLO BADLAPUR", "FLO GURGAON", "FLO BANGALORE", "FLO HYDERABAD 2 ", "FLO CHENNAI", "FLO LUCKNOW"]
facility_labels = ["Sakinaka", "Vasai", "Badlapur", "Gurgaon", "Bangalore", "Hyderabad", "Chennai", "Lucknow"]

# Initialize the data for Table 7
table_7_data = []

# Calculate the dispatch counts for each date and facility, considering only "Mattress" category
for date in dates:
    row_data = [date.strftime('%Y-%m-%d'), date.strftime('%A')]  # Add Date and Day of the week
    
    if date <= report_date:  # Only process dates up to the report date
        for facility in facility_names:
            count = data[
                (data['Dispatch Date'].dt.date == date.date()) & 
                (data['Facility'] == facility) & 
                (data['Category'] == 'Mattress') & 
                (~data['Tracking Number'].str.contains('CAN', na=False))
            ].shape[0]
            row_data.append(count)
        row_data.append(sum(row_data[2:]))  # Sum across all facilities for this date
    else:
        # For dates after the report date, append zeros
        row_data.extend([0] * len(facility_names))
        row_data.append(0)  # Sum across all facilities for this date will also be 0
    
    table_7_data.append(row_data)

# Convert to DataFrame and calculate the grand totals
table_7_df = pd.DataFrame(table_7_data, columns=['Date', 'Day'] + facility_labels + ['All Facilities'])

# Calculate the grand total for each facility
grand_total = ['Grand Total', '']
grand_total += table_7_df[facility_labels].sum().tolist()
grand_total.append(sum(grand_total[2:]))

# Append the grand total row
table_7_df.loc[len(table_7_df)] = grand_total

# Generate the output file name with the report date
output_file = f'Target Report [{report_date.strftime("%Y-%m-%d")}].xlsx'

# Load the existing Excel file to add Table 7
wb = load_workbook(output_file)
ws = wb.active

# Define the fill colors and font styles
orange_fill = PatternFill(start_color="FFA500", end_color="FFA500", fill_type="solid")
yellow_fill = PatternFill(start_color="FFD966", end_color="FFD966", fill_type="solid")
bold_font = Font(bold=True)

# Write the headers for Table 7 with formatting, starting at A74
headers = ['Date', 'Day'] + facility_labels + ['All Facilities']
for col_idx, header in enumerate(headers, start=1):
    cell = ws.cell(row=74, column=col_idx, value=header)
    cell.fill = orange_fill
    cell.font = bold_font
    cell.alignment = Alignment(horizontal='center')  # Center align the headers
    cell.border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

# Write the table data starting from cell A75
for r_idx, row in table_7_df.iterrows():
    for c_idx, value in enumerate(row, start=1):
        cell = ws.cell(row=75 + r_idx, column=c_idx, value=value)
        cell.alignment = Alignment(horizontal='center')  # Center align the data
        cell.border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

        # Apply yellow fill to the first 2 columns and last column (except the first and last rows)
        if r_idx != len(table_7_df) - 1:  # Not the last row
            if c_idx in [1, 2, len(headers)]:  # First 2 columns and last column
                cell.fill = yellow_fill

# Merge the last row's Column 1 and 2 and set the value for the merged cell
ws.merge_cells(start_row=74 + len(table_7_df), start_column=1, end_row=74 + len(table_7_df), end_column=2)
cell = ws.cell(row=74 + len(table_7_df), column=1)
cell.value = "Grand Total"
cell.fill = orange_fill
cell.font = bold_font
cell.alignment = Alignment(horizontal='center')  # Center align the Grand Total
cell.border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

# Apply formatting to the rest of the last row
for col_idx in range(3, 12):
    cell = ws.cell(row=74 + len(table_7_df), column=col_idx)
    cell.fill = orange_fill
    cell.font = bold_font
    cell.alignment = Alignment(horizontal='center')  # Center align the Grand Total row
    cell.border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

# Save the changes
wb.save(output_file)

print(f"Seventh table added successfully to {output_file} starting at A74 with dates, dispatch counts, grand totals, and specific formatting.")
```
